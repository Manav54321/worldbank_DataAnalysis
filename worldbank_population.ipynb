```python
df2 = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_408800.csv")
```

```python
df2.head()
```

```python
# Merge the two dataframes:
df_merged = pd.merge(df, df2, left_on='Country Name', right_on='TableName')

income_group_population = df_merged.groupby('IncomeGroup')['Total Population'].sum()

plt.figure(figsize=(10, 6))
income_group_population.plot(kind='bar', color=["red",  "orange", "gold", "limegreen"])
plt.xlabel('Income Group')
plt.ylabel('Total Population')
plt.title('Total Population by Income Group')
plt.show()
```

```python
df_1960 = df_merged[['Region', '1960']]
df_2022 = df_merged[['Region', '2022']]

region_pop_1960 = df_1960.groupby('Region')['1960'].sum()
region_pop_2022 = df_2022.groupby('Region')['2022'].sum()

# subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# 1960 pie chart
region_pop_1960.plot(kind='pie', autopct='%1.1f%%', ax=ax1, labels=None)
ax1.set_title('Distribution of Regions in 1960')
ax1.legend(region_pop_1960.index, loc='center left', bbox_to_anchor=(1, 0.5))

# 2022 pie chart
region_pop_2022.plot(kind='pie', autopct='%1.1f%%', ax=ax2, labels=None)
ax2.set_title('Distribution of Regions in 2022')
ax2.legend(region_pop_2022.index, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
# ----->
# - 01. plt.tight_layout() adjusts the spacing between subplots to prevent overlapping of labels, titles, and axes.
# - 02. This function automatically calculates the optimal spacing to ensure that all elements within the plot are clearly visible.
plt.show()
```

```python
income_group_1960 = df_merged.groupby('IncomeGroup')['1960'].sum()
income_group_2022 = df_merged.groupby('IncomeGroup')['2022'].sum()

fig, ax = plt.subplots(figsize=(10, 6))
income_group_1960.plot(kind='bar', ax=ax, position=0, width=0.4, label='1960', color="red")
income_group_2022.plot(kind='bar', ax=ax, position=1, width=0.4, label='2022', color="orange")

ax.set_xlabel('Income Group')
ax.set_ylabel('Total Population')
ax.set_title('Total Population by Income Group (1960 vs 2022)')
plt.legend()
plt.show()
```

```python
income_group_counts = df_merged['IncomeGroup'].value_counts()

plt.figure(figsize=(10, 6))
income_group_counts.plot(kind='bar', color = ["limegreen", "red",  "orange", "gold",])
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.title('Distribution of Income Groups')
plt.show()
```

```python
df_transposed = df_merged.set_index('Country Name').transpose()

years = df_transposed.loc['1960':'2022']

plt.figure(figsize=(15, 8))
plt.plot(years.index, years['World'], label='World') 
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Population in billion')
plt.title('World Population Growth from 1960 to 2022')
plt.legend()
plt.show()
```

```python
df_transposed = df_merged.set_index('Country Name').transpose()

years = df_transposed.loc['1960':'2022']

countries = ['India', 'China', 'United States', 'Germany']

plt.figure(figsize=(15, 8))
for country in countries:
    if country in years.columns:
        plt.plot(years.index, years[country], label=country)
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Population in billion')
plt.title('Population Growth from 1960 to 2022')
plt.legend()
plt.show()
```
